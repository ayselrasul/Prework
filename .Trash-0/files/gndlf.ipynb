```python
# DO NOT CHANGE THE FUNCTION SIGNATURE IN THE SOLUTION AREA

# Which sorcerer won the battle - Gandalf or Saruman?

import numpy as np

power = {
      'Fireball': 50, 
      'Lightning bolt': 40, 
      'Magic arrow': 10, 
      'Black Tentacles': 25,  
      'Contagion': 45
    }
    
gandalf = ['Fireball', 'Lightning bolt', 'Lightning bolt', 'Magic arrow', 'Fireball', 'Magic arrow', 'Lightning bolt', 'Fireball', 'Fireball', 'Fireball']

saruman = ['Contagion', 'Contagion', 'Black Tentacles', 'Fireball', 'Black Tentacles', 'Lightning bolt', 'Magic arrow', 'Contagion', 'Magic arrow', 'Magic arrow']

powers_gandalf = []
for weapon in gandalf:
  power_of_weapon = power[weapon]
  powers_gandalf.append(power_of_weapon)

powers_saruman = []
for weapon in saruman:
  power_of_weapon = power[weapon]
  powers_saruman.append(power_of_weapon)

def battle():
  number_of_rounds = len(gandalf)
  consecutive_win_gandalf = 0
  consecutive_win_saruman = 0
  for round_ in range(number_of_rounds):
    power_gandalf = powers_gandalf[round_]
    power_saruman = powers_saruman[round_]
    
    if power_gandalf > power_saruman:
      consecutive_win_gandalf += 1
      consecutive_win_saruman = 0
    elif power_gandalf < power_saruman:
      consecutive_win_saruman += 1
      consecutive_win_gandalf = 0
    
    if consecutive_win_gandalf == 3:
      return 'Gandalf'
    elif consecutive_win_saruman == 3:
      return 'Saruman'
  
#  Average of the spells in the lists? Round off your result to one decimal place.
def avg_spells():
    average_power = np.mean(powers_gandalf)
    return np.round(average_power, 1)
  
#  Standard deviation of the spells in the lists? Round off your result to one decimal place.
def stdev_spells():
    stdev_power = np.std(powers_gandalf)
    return np.round(stdev_power, 1)
```
