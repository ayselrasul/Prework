```python
#X-y split

y = df1['Total Claim Amount']  #we select a column which we define as target
X = df1.drop(['Total Claim Amount'], axis=1)  #we define the input with dataframe content except the target
```

```python
X_num = df1.select_dtypes(include=np.number)
X_cat = df1.select_dtypes(include=np.object)
```

```python
X_cat
```

```python
#Normalizing numerical data

from sklearn.preprocessing import Normalizer

transformer=Normalizer().fit(X_num)
X_normalized=transformer.transform(X_num)
print(X_normalized)
```

```python
sns.displot(X_num,bins=100)
plt.show()
sns.displot(X_normalized,bins=100)
plt.show()
```

```python
#Standartization

from sklearn.preprocessing import StandardScaler
transformer=StandardScaler().fit(X_num)
X_standard=transformer.transform(X_num)
print(X_standard)

```

```python
sns.distplot(X_num,bins=100)
plt.show()
sns.distplot(X_standard,bins=100)
plt.show()
```

```python
#One Hot/Label Encoding (categorical)

from sklearn.preprocessing import OneHotEncoder

X_cat_data = pd.get_dummies(categorical, drop_first=True)

ohe = OneHotEncoder(drop='first')
ohe.fit(X_cat_data)

X_ohe = ohe.transform(X_cat_data).toarray()
X_ohe.shape
```

```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
tr_cols=[]
for col in X_cat.columns:
    le.fit(X_cat[col])
    tr_cols.append(le.transform(X_cat[col]))
    
```

```python
#Concat DataFrames

X = np.concatenate([X_normalized, X_ohe], axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```
